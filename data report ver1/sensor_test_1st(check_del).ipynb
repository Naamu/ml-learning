```python
df = pd.DataFrame({'Sensor2':rawData1['sensor2'].values}, index=rawData1.index)
df.head()
```

```python
from scipy.fftpack import rfft
from scipy.fftpack import fftshift
from scipy.fftpack import fft
import numpy as np
```

```python
# FFR 
#t = np.linspace(-2*np.pi, 2*np.pi, len(df['Sensor2']))
#mid = np.ptp(df['Sensor2'])/2
#sine = mid + mid * np.sin(np.sin(t))
#sin_fft = np.abs(fftshift(rfft(sine)))
```

```python
# FFT 
sensor_fft = np.abs(fftshift(rfft(df['Sensor2'])))
```

```python
yf = fftshift(df['Sensor2'])
```

```python
plt.plot(yf)
```

```python
df['Sensor2_FFT']= sensor_fft
```

```python
print ("index of max FFT", np.argsort(sensor_fft)[-5:])
```

```python
#FFT  
plt.figure(figsize=(20,10))
plt.subplot(311)
plt.grid(True)
plt.plot(df['Sensor2'], label="Sensor2", color='r')
#plt.plot(sine, label="Sine")
#plt.axis([13,33, 0, 25])
plt.legend(loc='upper right')
plt.subplot(312)
plt.grid(True)
plt.plot(df['Sensor2_FFT'], label="Fftshift+rfft Sensor2")
plt.legend()
plt.subplot(313)
plt.grid(True)
plt.plot(yf, label="Fftshift Sensor2")
plt.legend()
```

```python
#     
plt.figure(figsize=(20,7))
plt.subplot(211)
plt.plot(df['Sensor2'], label="Sensor2", color='r')
plt.xlim(23.25,24.5)
#plt.plot(sine, label="Sine")
#plt.axis([13,33, 0, 25])
plt.grid(True)
plt.legend(loc='upper right')
plt.subplot(212)
plt.plot(df['Sensor2_FFT'], lw=1, label="Transformed Sensor2")
plt.xlim(23.25,24.5)
plt.grid(True)
plt.legend()
```

```python
from scipy.signal import medfilt
from scipy.signal import wiener
from scipy.signal import detrend
#     . medfilt()        .
#    . g(t)  s(t)   s*[s+n](t) . wiener() .
#     .
#   .        . detred() .
```

```python
df.head()
```

```python
Sensor2_medFT = medfilt(df['Sensor2'], 15)
Sensor2_wieFT = wiener(df['Sensor2'], 15)
Sensor2_detFT = detrend(df['Sensor2'])
```

```python
df['Sensor2_medFT'] = Sensor2_medFT
df['Sensor2_wieFT'] = Sensor2_wieFT
df['Sensor2_detFT'] = Sensor2_detFT
```

```python
df.head()
```

```python
#
#plt.plot(df['Sensor2'])
plt.figure(figsize=(20,7))
plt.plot(df.Sensor2, label='Origin')
plt.plot(df.Sensor2_medFT, lw=2,label="median")
plt.plot(df.Sensor2_wieFT, '--', label="wiener")
plt.plot(df.Sensor2_detFT, label="detrend")
plt.xlabel("Time")
plt.legend()
plt.show()
```
