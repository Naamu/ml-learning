```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sensor_data_jung = pd.read_csv("data/data_namuk/sensor_jung_all.csv")
sensor_data_jung = sensor_data_jung.rename(columns = {'Unnamed: 0': 'index'})
sensor_data_jung = sensor_data_jung.set_index('index')
```

```python
sensor_data_jung.describe()
```

```python
sensor_data_gun = pd.read_csv("data/data_namuk/sensor_gun_all.csv")
sensor_data_gun = sensor_data_gun.rename(columns = {'Unnamed: 0': 'index'})
sensor_data_gun = sensor_data_gun.set_index('index')
```

```python
sensor_data_gun.describe()
```

```python
sensor_data_all = pd.concat([sensor_data_gun, sensor_data_jung], ignore_index=True)
sensor_data_all.describe()
```

```python
print(sensor_data_all['label'].value_counts())
```

```python
sensor_data_all.to_csv("data/data_namuk/sensor_data_all.csv", mode='w')
```
