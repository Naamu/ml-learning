```python
df = pd.DataFrame({'Sensor2':rawData1['sensor2'].values}, index=rawData1.index)
df.head()
```

```python
from scipy.fftpack import rfft
from scipy.fftpack import fftshift
from scipy.fftpack import fft
import numpy as np
```

```python
#싸인함수에 FFR 적용하기
#t = np.linspace(-2*np.pi, 2*np.pi, len(df['Sensor2']))
#mid = np.ptp(df['Sensor2'])/2
#sine = mid + mid * np.sin(np.sin(t))
#sin_fft = np.abs(fftshift(rfft(sine)))
```

```python
#센서에 FFT 적용하기
sensor_fft = np.abs(fftshift(rfft(df['Sensor2'])))
```

```python
yf = fftshift(df['Sensor2'])
```

```python
plt.plot(yf)
```

```python
df['Sensor2_FFT']= sensor_fft
```

```python
print ("index of max FFT", np.argsort(sensor_fft)[-5:])
```

```python
#FFT로 전체 나타내기
plt.figure(figsize=(20,10))
plt.subplot(311)
plt.grid(True)
plt.plot(df['Sensor2'], label="Sensor2", color='r')
#plt.plot(sine, label="Sine")
#plt.axis([13,33, 0, 25])
plt.legend(loc='upper right')
plt.subplot(312)
plt.grid(True)
plt.plot(df['Sensor2_FFT'], label="Fftshift+rfft Sensor2")
plt.legend()
plt.subplot(313)
plt.grid(True)
plt.plot(yf, label="Fftshift Sensor2")
plt.legend()
```

```python
# 두개로 표 나타내고 일부분 잘래내기
plt.figure(figsize=(20,7))
plt.subplot(211)
plt.plot(df['Sensor2'], label="Sensor2", color='r')
plt.xlim(23.25,24.5)
#plt.plot(sine, label="Sine")
#plt.axis([13,33, 0, 25])
plt.grid(True)
plt.legend(loc='upper right')
plt.subplot(212)
plt.plot(df['Sensor2_FFT'], lw=1, label="Transformed Sensor2")
plt.xlim(23.25,24.5)
plt.grid(True)
plt.legend()
```

```python
fig, ax1 = plt.subplots(figsize=(20,7))
ax2 = ax1.twinx()
line1 = ax1.plot(df.Sensor2, color='b', label = 'Sensor2') #linestyle='--'
line2 = ax2.plot(df.Sensor2_FFT, color='r', label = 'Sensor2_FFT')
lines = line1 +line2
labels=[l.get_label() for l in lines]
plt.legend(lines, labels)
plt.grid(True)
```

```python
from scipy.signal import medfilt
from scipy.signal import wiener
from scipy.signal import detrend
#중앙값 필터는 현재 데이터의 중앙값을 계산한다. medfilt()함수를 통해 윈도우 사이즈를 파라미터로 해서 사용할 수 있다.
#워너필터는 통계치를 사용해 노이즈를 제거한다. g(t) 필터와 s(t)신호에서 출력은 중첩된 s*[s+n](t)형태로 나타난다. wiener()함수를 사용하면된다.
#해당함수또한 윈도우 크기를 파라미터로 지정할수 있다.
#디트렌드 필터는 트렌드를 제거한다. 해당 트렌드는 선형일 수 있고 상수일 수 있다. detred()함수를 사용하면된다.
```

```python
df.head()
```

```python
Sensor2_medFT = medfilt(df['Sensor2'], 15)
Sensor2_wieFT = wiener(df['Sensor2'], 15)
Sensor2_detFT = detrend(df['Sensor2'])
```

```python
df['Sensor2_medFT'] = Sensor2_medFT
df['Sensor2_wieFT'] = Sensor2_wieFT
df['Sensor2_detFT'] = Sensor2_detFT
```

```python
df.head()
```

```python
#필터링
#plt.plot(df['Sensor2'])
plt.figure(figsize=(20,7))
plt.plot(df.Sensor2, label='Origin')
plt.plot(df.Sensor2_medFT, lw=2,label="median")
plt.plot(df.Sensor2_wieFT, '--', label="wiener")
plt.plot(df.Sensor2_detFT, label="detrend")
plt.xlabel("Time")
plt.legend()
plt.show()
```
